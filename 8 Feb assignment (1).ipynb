```python
"""Ans1: Abstract class is a blue print for other classes. It helps us to set of methods that must be created within any child classes built from abstract class.


Example:"""

import abc
class abst:
    @abc.abstractmethod
    def name(self):
        pass
    
    @abc.abstractmethod
    def rate(self):
        pass
    
class medical(abst):
    def name(self):
        return "PCM"
    def rate(self):
        return 20

details=medical()
details.name()
    
```

```python
"""Ans4: We can achieve data abstraction by creating abstract class.
First we have to import abc module then just above the function write @abc.abstractmethod() for making the blue print, then we have to abstract this class in another for using that blueprint and define the method according to our use.
Example:"""
import abc
class abst:
    @abc.abstractmethod
    def name(self):
        pass
    
    @abc.abstractmethod
    def rate(self):
        pass
    
class medical(abst):
    def name(self):
        return "PCM"
    def rate(self):
        return 20

details=medical()
details.name()
    
```
